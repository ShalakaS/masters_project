```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
# show all columns on pandas dataframes
pd.set_option('display.max_columns', None)
# see numbers after decimal points
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
url = "https://dashboard.doj.gov.in/sanctiondata/index.php"
```

```python
response1 = requests.get(url).content
response1
```

```python
soup_doc = BeautifulSoup(response1, 'html.parser')
soup_doc
```

```python
all_scripts = soup_doc.find_all("script")
```

```python
all_text = all_scripts[30].get_text()
all_text
```

```python
labels = re.findall(r'label\s*:\s*\'([^"]+)\',', all_text)
sanctioned_strength = re.findall(r'"Sanctioned Strength"\s*:\s*([0-9]+),', all_text)
working_strength = re.findall(r'"Working Strength"\s*:\s*([0-9]+),', all_text)
vacancy_strength = re.findall(r'"Vacancy Strength"\s*:\s*([0-9]+),', all_text)
```

```python
len(labels)
```

```python
#HAVE TO ADD NAGALAND SEPARATELY BECAUSE "NAGALAND" LABELS DID NOT GET SCRAPED. BECAUSE REGEX IS ANNOYING!!
#SOURCE: https://www.tutorialspoint.com/How-to-insert-an-object-in-a-list-at-a-given-position-in-Python#:~:text=Algorithm%20(Steps),-Following%20are%20the&text=the%20input%20list-,Enter%20the%20item%20to%20be%20inserted%20into%20the%20list%20and,inserted%20as%20arguments%20to%20it.

# input list
labels

# giving the item to be inserted
insertItem = "Nagaland"

# giving the index value at which the item to be inserted
indexValue = 23

# inserting the given list item at the specified index(here 2)
labels.insert(indexValue, insertItem)

# printing the list after insertion
labels[23]
```

```python
df = pd.DataFrame({"sanctioned_strength": sanctioned_strength,
             "working_strength": working_strength,
             "vacancy": vacancy_strength,
            "states": labels})
df
```

```python
df['vacancy'] = df['vacancy'].astype(int)
df['sanctioned_strength'] = df['sanctioned_strength'].astype(int)
df['working_strength'] = df['working_strength'].astype(int)
```

```python
df.head(1)
```

```python
df['vacancy_pct'] = df['vacancy']/df['sanctioned_strength'] * 100
```

```python
df.head(2)
```

```python
df.columns = df.columns.str.replace('states', 'State')
```

```python
df['State'] = df['State'].str.replace('Chhatisgarh', 'Chhattisgarh').str.replace(" and ", " And ").str.replace("Dadra And Nagar Haveli", "The Dadra And Nagar Haveli And Daman And Diu").str.replace("Daman And Diu", "The Dadra And Nagar Haveli And Daman And Diu")
```

```python
df['State'] = df['State'].str.replace('The Dadra And Nagar Haveli And Daman And Diu', "_").str.replace('The Dadra And Nagar Haveli And', "")
```

```python
df['State'] = df['State'].str.replace("_", "The Dadra And Nagar Haveli And Daman And Diu")
```

```python
df.sort_values(by='vacancy_pct', ascending=False)
```

```python
df['vacancy_pct'][1:].mean().round()
```

```python
df1 = pd.read_csv('NDAP_API_data.csv')
df1.head(2)
```

```python
df2 = df1.groupby('State').total_pending.sum().reset_index()
df2.head(2)
```

```python
df2['State'] = df2['State'].str.replace(" Islands", "")
```

```python
df3 = df1.groupby('State').total_criminal.sum().reset_index()
df3.head(2)
```

```python
df3['State'] = df2['State'].str.replace(" Islands", "")
df3.columns = df3.columns.str.replace("State", "States")
```

```python
df_new = pd.concat([df2, df3], axis=1)
df_new = df_new.drop(columns=['Statess', 'States'])
```

```python
df_final = df.merge(df_new, on='State', how='right')[1:].reset_index(drop=True)
```

```python
df_final['caseload_per_judge'] = df_final['total_pending'] / df_final['working_strength']
```

```python
df_final.sort_values(by='caseload_per_judge', ascending=False).to_csv('caseload_per_judge.csv', index=False)
```
